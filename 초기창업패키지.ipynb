## 1. Problem(문제인식)

```python
Problem_template = """
1. 문제인식(Problem)
1) 창업 아이템의 개발 필요성
    - 현재 시장에서 해결되지 않은 주요 문제점과 해당 문제는 국내·외에서 어떤 영향을 미치는지 작성하세요.
    - 현재 시장에서 사용되는 기존 해결책(경쟁 제품/서비스)의 한계점.
    - 창업 아이템이 이러한 문제를 어떻게 해결할 것인지 작성하세요.
    - 소비자 혹은 타겟 고객은 이 문제에 대해 어떤 불편이 있는지 작성하세요.
    - 기존 시장에서 유사한 해결책이 있었던 경우와 어떤 부분에서 부족했는지 작성하세요.
예시) 2) 위에 내용과 알맞는 항목을 만들어 작성하세요.
    """
```

```python
prompt1 = f"""
당신은 초기창업자를 위한 사업계획서 전문가입니다.
논리적이고 설득력 있는 사업계획서를 작성하세요.
문장은 간결하게, 핵심을 강조하며, 투자자 및 심사위원이 주목할 수 있도록 수치, 데이터, 시장 반응을 명확히 표현하세요.
이 작업은 창업자의 생계와 미래가 걸린 **절대적으로 중요한 프로젝트**입니다.
작성된 사업계획서의 완성도가 곧 **합격과 실패를 결정**하며, 이를 통해 창업자는 사업을 현실화하고 생존할 기회를 얻게 됩니다.
초안을 **작성할 때는 {loaded_data['문제인식']} 기반으로만** 쓰며, {Problem_template}에 내용들을 참고하여 순서에 맞게 작성해주세요.

**반드시 지켜야 할 출력 규칙**
    1. 절대 #, * 기호 및 마크다운 스타일 절대 금지
    2. 항목 표시 시 "1)", "2)" 형식만 사용 (숫자+닫는 괄호)
    3. 데이터 출처 명시 : 모든 통계 수치에 출처를 포함. (ex. 출처: 통계청)
"""

completion1 = client.chat.completions.create(model='gpt-4o',
                                             messages=[{'role':'user', 'content':prompt1}],
                                             temperature=0.7,
                                             presence_penalty=0.8,
                                             frequency_penalty=0.5)
print(completion1.choices[0].message.content)
```
